```r

metadf <- read.csv('data_all.csv')
glimpse(metadf)
```

```r
# AUTHOR に "UNKNOWN" を含む行を除去（大文字小文字無視）
df_no_unknown <- metadf %>%
  mutate(AUTHOR = stringr::str_squish(AUTHOR)) %>%  # 前後/連続空白を整理
  filter(!stringr::str_detect(stringr::str_to_upper(coalesce(AUTHOR, "")),"UNKNOWN"))

# 確認
# df_no_unknown %>% count(AUTHOR) %>% arrange(desc(n))

df_unique_only <- df_no_unknown %>%
  mutate(
    AUTHOR_trim = stringr::str_to_upper(stringr::str_squish(AUTHOR)),
    TITLE_trim  = stringr::str_to_upper(stringr::str_squish(TITLE))
  ) %>%
  add_count(AUTHOR_trim, TITLE_trim) %>%
  filter(n == 1) %>%
  select(-AUTHOR_trim, -TITLE_trim, -n)

# 確認
# df_unique_only %>% count(AUTHOR) %>% filter(n>1) %>% arrange(desc(n))

# ★ AUTHOR が1件しかない作者を除外（AUTHORごとに件数カウント→2件以上だけ残す）
df_author_ge2 <- df_unique_only %>%
  add_count(AUTHOR, name = "n_author") %>%
  filter(n_author >= 2) %>%
  select(-n_author)

df_author_ge2 %<>% select(filename,Condition,TITLE,AUTHOR,LOCATION)

# 必要なら保存
# readr::write_csv(df_author_ge2, "data_for_custom_search.csv")
```
