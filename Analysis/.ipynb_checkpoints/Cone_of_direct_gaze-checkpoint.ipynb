# The face and gaze direction during the Renaissance

## Overview

We investigated the orientation of faces in portrait paintings from the Renaissance period (14th to late 16th century) to understand artistic trends in subject depiction. Using OpenFace (version 2.2.0), we analyzed 1,017 portraits from the Web Gallery of Art (as of December 2024). Among these, 871 faces were successfully processed, including those from multi-figure compositions (which were then left out of the analysis). The results revealed a bimodal distribution of facial orientations, symmetrically balanced to the left and right of the frontal axis. Focusing on faces turned to the right from a frontal view, we calculated the mean rotation angle to be approximately 23° from the frontal position. We detail the results below.

## Create variable with cone of direct gaze

```r
# ロジスティック関数で閾値を計算する関数
threshold_func <- function(yaw, max_th = 7.5, min_th = 0.5, k = 0.08, midpoint = 30) {
  min_th + (max_th - min_th) / (1 + exp(k * (yaw - midpoint)))
}

metadf <- metadf %>%
  rowwise() %>%
  mutate(
    # 顔の向きに応じた閾値を計算
    threshold = threshold_func(abs(headYaw)),
    
    # 直視か逸らしを判定
    gaze_cat = ifelse(abs(gazeYaw) <= threshold, "direct", "avert")
  ) %>%
  ungroup()

head(metadf)
```

```r
metadf %>% group_by(TF,gaze_cat,Condition) %>% 
    filter(gazeYaw<10,gazePitch<40,headYaw>-15,Condition!='Post_Renaissance') %>%
    summarise(n=n()) %>% ggplot(aes(y=n,x=as.factor(TF),fill=gaze_cat)) + 
    geom_bar(position='dodge',stat='identity') + 
    facet_wrap(.~Condition) + theme(legend.position = 'bottom') + 
    labs(x='Timeframe',y='Number of Paintings',fill='Gaze Shift') + 
    scale_fill_manual(values=c('avert'='#E69F00','straight'='#56B4E9'))
# metadf %>% group_by(TF,gaze_cat,Condition) %>% summarise(n=n())

fit2 <- metadf %>% mutate(gaze_cat2=ifelse(gaze_cat=='avert',0,1)) %>% filter(Condition!='Post_Renaissance', date>1400, date<1600,SCHOOL == 'Flemish' | SCHOOL == 'German' | SCHOOL == 'Italian') %>% glm(gaze_cat2~date*Condition+SCHOOL, family = binomial("logit"),data=.)
summary(fit2)
glm_p <- ggpredict(fit2, terms = c("date","Condition"))
# glm_p
(p <-ggplot() + 
     geom_line(data = glm_p, aes(x = x, y = predicted, color = group), show.legend = F) + 
     geom_ribbon(data = glm_p, aes(x = x, ymin = conf.low, ymax = conf.high, fill = group), alpha = 0.3) +
     ylab('Proportion of direct gaze (1=direct)') + xlab('Year') + scale_x_continuous(breaks=seq(1400,1650,50)))
# ggsave('Gaze_Shift_Left_Renai_only.png',p)

emm.result <- emmeans(fit2, ~ SCHOOL, at = list(date=1400),pbkrtest.limit = 4560)
summary(emm.result, infer=TRUE)
pairs(emm.result,pbkrtest.limit = 4560)

emm.gen <- emtrends(fit2, "SCHOOL", var = "date",pbkrtest.limit = 4560)
summary(emm.gen, infer=TRUE)
pairs(emm.gen,pbkrtest.limit = 4560)
```

## Gaze during the Renaissance

### From Right to Center gaze (only Renaissance paintings)

We conducted a linear regression to examine the temporal trend in gaze direction of painted figures. The dependent variable was horizontal gaze angle, and the predictor was the year of creation. Since avert gaze could mean from looking toward the right (indicated by negative degrees) to looking toward the left (indicated by positive degrees), we first analysed the gaze ranging from the right to direct gaze. Since humans perceive direct gaze in the area from -10 to 10 degrees, which is known as the cone of direct gaze, we analysed from -90 degrees to 10 degrees. Figure X shows the overall pattern between the beginning and the end of the Renaissance. A clear trend could be seen from the avert gaze (indicated by negative degree in horizontal gaze) to the direct gaze (indicated by data around the cone of direct gaze).

The analysis showed a significant positive effect of date ($\beta = 0.065$, $p < .001$), suggesting that gaze direction shifted from avert to direct over time. Earlier paintings tended to feature avert gazes, while later works increasingly depicted figures looking directly at the viewer. This supported the earlier result from Morin (2013).

```r
# head(metadf)
# metadf %>% filter(date>1400,date<1600,gazeYaw<10,headPitch<90,headPitch>-90,Condition=='Renaissance') %>% lmer(gazeYaw~date+(1|AUTHOR),data=.) -> fit1
# fit1 %>% summary()

metadf %>% filter(date>1400,date<1600,gazeYaw<10,headPitch<90,headPitch>-90,Condition=='Renaissance') %>% lm(gazeYaw~date,data=.) -> fit1
fit1 %>% summary()

glm_p <- ggpredict(fit1, terms = c("date"))

(p <-ggplot() + 
     geom_line(data = glm_p, aes(x = x, y = predicted), show.legend = F) + 
     scale_x_continuous(breaks=seq(1400,1600,50)) +
     geom_ribbon(data = glm_p, aes(x = x, ymin = conf.low, ymax = conf.high), alpha = 0.3,show.legend = F) +
     stat_summary(data = filter(metadf,date>1400,date<1600,gazeYaw<10,headPitch<90,headPitch>-90,Condition=='Renaissance'), aes(x = date, y = gazeYaw), fun= mean, geom="point", size = 5, shape = 18, position = position_dodge(width = .75)) +
     ylab('Horizontal gaze') + xlab('Year') + geom_hline(yintercept = c(-10, 10), linetype = "dotted", linewidth = 1.5))
# ggsave('Gaze_Shift_Left_Renai_only.png',p)
```

### From Left to Center gaze (only Renaissance paintings)

To test whether this trend could also be seen in leftward averted gaze, we conducted a separate analysis including only data from -10 degrees to +90 degrees (FigureX ). In this range, however, the regression did not yield a significant effect of date ($\beta = -0.020$, $p = .180$). While the intercept was positive ($\beta = 39.09$, $p = .09$), indicating that figures on average looked slightly to the left, though there was no reliable temporal trend. This indicated that there might have been a different cultural selection between leftward gaze and rightward gaze.

## Left-cheek bias

```r

metadf2 <- metadf %>% filter(Condition=='Renaissance') %>% mutate(headdirect=ifelse(headYaw>7.5,'left_cheek',ifelse(headYaw< -7.5,'right_cheek','straight')))

(chisq <- metadf2 %>% group_by(headdirect) %>% summarise(n=n()))


# χ²検定
chi <- chisq.test(x=chisq$n)
chi

# Left vs Right
p1 <- binom.test(x = chisq$n[1], 
                 n = chisq$n[1] + chisq$n[2], 
                 p = 0.5)$p.value

# Left vs Straight
p2 <- binom.test(x = chisq$n[1], 
                 n = chisq$n[1] + chisq$n[3], 
                 p = 0.5)$p.value

# 多重比較補正（Bonferroni）
p_adj <- p.adjust(c(p1, p2), method = "bonferroni")
names(p_adj) <- c("Left vs Right", "Left vs Straight")
p_adj
```

### multi-log regression

```r

# 1) データ整形：Left/Right/Straight に分類（±7.5°閾値）
dat <- metadf %>% filter(Condition=='Renaissance') %>%
  mutate(
    headdirect = case_when(
      headYaw >  7.5 ~ "left_cheek",
      headYaw < -7.5 ~ "right_cheek",
      TRUE           ~ "straight"
    ),
    headdirect = factor(headdirect, levels = c("left_cheek","right_cheek","straight")),
    date_sc = scale(date)[,1]   # 年代を標準化
  ) %>%
  filter(!is.na(headdirect), !is.na(date_sc))

# 2) 多項ロジスティック回帰（固定効果）
#    参照カテゴリは left_cheek（= Left Cheek Bias が基準）
m_mn <- multinom(headdirect ~ date_sc, data = dat, trace = FALSE)

summary(m_mn)
broom::tidy(m_mn, conf.int = TRUE, exponentiate = TRUE)  # 相対オッズ比で確認

# 3) 「Left Cheek Biasが年代とともに弱まるか？」の検定
#   → left を基準にした right/straight の相対オッズに対する date_sc の傾きが正なら
#     時代が進むほど「right or straight」相対比が増え、Left優位が弱まる解釈
#   emmeans でトレンド（傾き）を直接テスト
em_trend_right  <- emtrends(m_mn, ~ 1, var = "date_sc", mode = "latent", which = "right_cheek")
em_trend_straight <- emtrends(m_mn, ~ 1, var = "date_sc", mode = "latent", which = "straight")
em_trend_right
em_trend_straight

# 4) 予測確率の連続曲線を可視化（年代 vs 各カテゴリ確率）
grid <- tibble(
  date = seq(min(dat$date), max(dat$date), length.out = 200)
) %>%
  mutate(date_sc = (date - mean(dat$date, na.rm=TRUE))/sd(dat$date, na.rm=TRUE))

emm_curve <- emmeans(m_mn, ~ headdirect | date_sc, at = list(date_sc = grid$date_sc), type = "response")
plotdf <- as.data.frame(emm_curve) %>%
  mutate(date = grid$date[match(date_sc, grid$date_sc)])
ggplot(plotdf, aes(x = date, y = prob, color = headdirect)) +
  geom_line(size = 1) +
  # scale_y_continuous(labels = percent_format(accuracy = 1)) +
  labs(x = "Year", y = "Predicted probability", color = "Head orientation",
       title = "Predicted probabilities of head orientation over time") +
  theme_minimal(base_size = 13)
```
